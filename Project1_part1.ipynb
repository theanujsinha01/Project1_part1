PROBLEM OBJECTIVE: TOP CUISINES
------------------------------
(1)DETERMINE THE TOP THREE MOST COMMON CUISINES IN THE DATASET.

(2)CALCULATE THE PERCENTAGE OF RESTAURANTS THAT SERV EACH OF THE TOP CUISINES.


```python
#top 3 most common cuisines
print("TOP THREE COMMON CUISINES")
top_cuisine = data.Cuisines.value_counts()
top3cuisines=top_cuisine.head(3)
top3cuisines
```

```python
plt.figure(figsize=(6,6))
colors = ['#1A5319', '#508D4E','#80AF81']
plt.bar(top3cuisines.index, top3cuisines , color=colors)
plt.title('TOP THREE COMMON CUISINES')
plt.show()
```
